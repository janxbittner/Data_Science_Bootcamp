Importing modules.

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(style = 'darkgrid',palette= 'viridis')
```

Create data set using random values from numpy module. Show first 3 rows.

```
df = pd.DataFrame(np.random.rand(100,100))
df.head(3)
```

Create heat map of dataframe.

```
sns.heatmap(df)
```

Import data set about flight from seaborn module. Show first 3 rows.

```
f_df = sns.load_dataset('flights')
f_df.head(n =3)
```

Show informations about columns.

```
f_df.info()
```

Show description of data set.

```
f_df.describe()
```

Create pivot table with numbers of passengers dependance of date.

```
p_t = pd.pivot(f_df,values = 'passengers',index = 'month',
                     columns = 'year')
```

Show first 3 rows of new pivot table.

```
p_t.head(3)
```

Create heatmap of numbers of passengers dependance of month and year.

```
sns.heatmap(p_t, annot = True, fmt = 'd')
```
