### <a name='10'></a> Exercise 250
Using the _cv2.rectangle () _ function from the _opencv_ library, add a rectangle border to the _img_ image surrounding the python logo. Set the color to green `color = (0, 255, 0)` and the thickness to 3 pixels.

```python
_= plt.imshow(cv2.rectangle(img= img.copy(),pt1 =(90,25),pt2 =(210,142),color = (0,255,0),thickness = 3))
```

Using the _cv2.circle () _ function from the _opencv_ library, add a circle to the _img_ image that covers the python logo. Set the color to green `color = (0, 255, 0)`.

```python
_= plt.imshow(cv2.circle(img= img.copy(), center = (150,82), radius = 60, color = (0,255,0),thickness = -1))
```

Using the _cv2.putText () _ function from the _opencv_ library, add the text 'Logo: Python' to the _img_ image in the upper left corner.
```python
_= plt.imshow(cv2.putText(img= img.copy(),text = 'Logo: Python', org = (10, 40),
                          fontFace = cv2.FONT_HERSHEY_SIMPLEX,
                          fontScale = 1.2, color = (255 ,255, 0),thickness = 5))
```
